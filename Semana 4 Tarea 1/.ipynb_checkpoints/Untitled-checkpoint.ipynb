Write down a code able to identify whether a pair of segments share points.
Three possible answers should be available: segments do not intersect, segments do intersect

Para este primer punto vamos a basarnos de la clase punto, la cual presenta un copnstructor básico y algunos métodos de visualización de los objetos. 
Tambien se implementó una clase Segment definida por objetos punto

```python
class Point:
    def __init__(self,cordinates,name):
        self.x = cordinates[0]
        self.y = cordinates[1]
        self.name = name
        
    def __repr__(self):
        return (f"({self.x}, {self.y})" )
    
    def __str__(self):
        return (self.name)
    
    def __sub__(self, other):
        if isinstance(other, Point):
            return Point((self.x - other.x, self.y - other.y), f"{self.name}-{other.name}")
        else:
            raise TypeError("Unsupported operand type for -: Point and {}".format(type(other).__name__))

    def __eq__(self, other):#definimos la igualdad
        if not isinstance(other, Point):
            return False
        return self.x == other.x and self.y == other.y
    

def cross_point(p0, p1, p2):
    prod1 = (p1.x - p0.x) * (p2.y - p0.y)
    prod2 = (p1.y - p0.y) * (p2.x - p0.x)
    res = prod1 - prod2
    if res < 0: 
        return -1  # p1 está en sentido horario respecto a p2 - p0
    elif res > 0:
        return 1   # p1 está en sentido anti-horario respecto a p2 - p0
    else:
        return 0   # los segmentos son colineales


def visualization(points,p=None):
    fig, ax = plt.subplots(figsize=(3,3))
    colors = ['blue','red','green','orange','purple','cyan','brown']
    
    for i in range(0, len(points)-1,2):
        plt.arrow(points[i].x, points[i].y, points[i+1].x - points[i].x, points[i+1].y - points[i].y, head_width =0.05, head_length=0.05,
                 color=colors[i], label= f"{points[i+1]}-{points[i]}")
        
    if p != None: plt.scatter(p.x, p.y)
    plt.legend()
    plt.show()
```

Vamos a introducir dos métodos adicionales para abordar este punto. En primer lugar, implementaremos **on_segment**, una función que nos permitirá determinar si un punto en particular pertenece a un segmento de línea. Específicamente, evaluaremos si el punto Pr se encuentra dentro del segmento generado por los puntos Pi y Pp.

El segundo método, **intersect_segment**, se inspira en un pseudocódigo que encontramos en el libro dos. Esta función desempeña un papel crucial en la detección de intersecciones entre segmentos. Funciona al identificar cambios de signo en los "ángulos" d0, d1, d2 y d3. La presencia de tales cambios de signo nos indica la posible intersección de segmentos. Luego de esto, atenderemos a los casos donde alguno de estos "ángulos" sea cero y verificaremos si el punto correspondiente se encuentra efectivamente en el segmento.

Para cerrar, implementaremos la función **found_point**. Una vez que se detecta una intersección, esta función se activa con el propósito de determinar el punto o intervalo en el cual se produce la intersección. se revisa si esa intersección es por algun punto extremo, un intervalo o un solo punto. EL intervalo se verifica con la función **interval_intersection**. Este último paso es fundamental para comprender la naturaleza y la localización precisa de la intersección entre los segmentos involucrados.

```python
def on_segment(p, q, r): # Pinicial, Pprincipal, Preferente(respecto a)
    if ( (q.x <= max(p.x, r.x)) and (q.x >= min(p.x, r.x)) and 
           (q.y <= max(p.y, r.y)) and (q.y >= min(p.y, r.y))):
        return True
    return False

def intesect_segments(P0, P1, P2, P3):
    d0 = cross_point(P2, P0, P3)
    d1 = cross_point(P2, P1, P3)
    d2 = cross_point(P0, P2, P1)
    d3 = cross_point(P0, P3, P1)

    if (d0 == 0 and d1 == 0 and d2 == 0 and d3 == 0):
        # Segmentos colineales
        if on_segment(P2, P0, P3) or on_segment(P3, P0, P1) or on_segment(P0, P2, P1) or on_segment(P1, P2, P3):
            return True  # Se intersectan en un punto final
        return False  # No se intersectan
    elif d0 * d1 <= 0 and d2 * d3 <= 0:
        return True
    else:
        return False


def found_point(P0, P1, P2, P3):
    if P0 == P2 or P0 == P3:   return P0
    elif P1 == P2 or P1 == P3: return P1
    segment1 = (P0, P1)
    segment2 = (P2, P3)
    
    if interval_intersection(segment1, segment2) is not None:
        return interval_intersection(segment1, segment2)

    else:

        x = np.array([[P0.x - P1.x, P3.x - P2.x], [P0.y - P1.y, P3.y - P2.y]])
        y = np.array([P3.x - P1.x, P3.y - P1.y])
        xinv = np.linalg.inv(x)
        res = np.dot(xinv, y)
        alpha = res[0]
        beta = res[1]
        Px_R = alpha * P0.x + (1 - alpha) * P1.x
        Py_R = alpha * P0.y + (1 - alpha) * P1.y
        Pres = Point((Px_R, Py_R), 'P_int')
        return Pres
    
        
def interval_intersection(segment1, segment2):
    min_x = max(min(segment1[0].x, segment1[1].x), min(segment2[0].x, segment2[1].x))
    max_x = min(max(segment1[0].x, segment1[1].x), max(segment2[0].x, segment2[1].x))
    min_y = max(min(segment1[0].y, segment1[1].y), min(segment2[0].y, segment2[1].y))
    max_y = min(max(segment1[0].y, segment1[1].y), max(segment2[0].y, segment2[1].y))
    
    if min_x <= max_x and min_y <= max_y:
        return Point((min_x, min_y), 'Interval Start'), Point((max_x, max_y), 'Interval End')
    else:
        return None
```

#### Empezamos definiendo una clase Point la cual contará con un método para coonvertir una lista de puntos en una lista de objetos Point **convertir_lista_points** , unos métodos de visualización y unos operadores para comparar objetos Point

```python
class Point:
    def __init__(self,cordinates,name):
        self.x = cordinates[0]
        self.y = cordinates[1]
        self.name = name
           
    def __repr__(self):
        return (f"({self.x}, {self.y})" )

    def __str__(self):
        return (self.name)
    
    def __lt__(self, other): #definimos el menor
        if not isinstance(other, Point):
            raise ValueError("Cannot compare Point with non-Point object")
        return (self.y > other.y or (self.y == other.y and self.x < other.x))
    
    def __eq__(self, other):#definimos la igualdad
        if not isinstance(other, Point):
            return False
        return self.x == other.x and self.y == other.y
    
    def __le__(self, other):
        if not isinstance(other, Point):
            raise ValueError("Cannot compare Point with non-Point object")
        return self.__lt__(other) or self.__eq__(other)
    
    def imprimir_puntos(points): 
        [print(p) for p in points]
    
    @classmethod
    def convertir_lista_points(cls, list_points):
        points = []
        for i, coords in enumerate(list_points):
            p = cls(coords,f'P{i}')
            points.append(p)
        return points
        
```

definimos un nuevo croos point (aunque es exactamente igual al anterior) la función **check_last3_poiints** el cual nos permitirá detectaar giros antihorarios dado 3 puntos. por último un método para visualizar la envolvente convexa y los puntos en el plano.

El método principal convex_hull funciona al calcular un pivote y un nex (next point) el next es aleatorio y se actualiza dependiendo de si se realizo un giro antihorario, si no encuentra más giros este valor se agrega a la envolvente convexa, actualizo el pivote a mi último punto de la envolvente y repito el procedimiento.

```python
def cross_product(p0,p1,p2):# tomamos p0 como el punto de origen, p1 como el punto que queremos categorizar, respecto a p2

    prod1 = (p1.x - p0.x) * (p2.y - p0.y)
    prod2 = (p1.y - p0.y) * (p2.x - p0.x)
    res = prod1 - prod2
    if res < 0 : return 1   #print("p1 esta en sentido horario respecto a p2 - p0")
    elif res > 0 :return -1 #print ("p1 esta en sentido anti-horario respecto a p2 - p0")
    else: return 0          #print(" los segmentos son colineales")


def check_last3_points(points): # detecta si se agregó algún punto que generó un giro antihorario
    pn = points[-1]
    pn_1 = points[-2]
    pn_2 = points[-3]
    cross = cross_product(pn_2,pn_1,pn)
    if cross == 1 or cross == 0:
        #print ("giro a la derecha (horario)")
        return 1
    else:
        #print ("giro a la izquierda (antihorario)")
        return 0
    

def visualization(points, convex):
    # Plotting the original points and their coordinates
    x = [coor.x for coor in points]
    y = [coor.y for coor in points]

    # Create a figure and axes
    fig, axs = plt.subplots(figsize=(12, 6))

    # Plot the original points and their x-coordinate labels in the left subplot
    axs.scatter(x, y, color='blue')

    # Add x-coordinate labels
    for point in points:
        x_label = point.name
        axs.text(point.x, point.y, x_label, ha='center', va='bottom')

    # Plot the lines connecting the points
    for i in range(len(convex) - 1):
        axs.plot([convex[i].x, convex[i+1].x], [convex[i].y, convex[i+1].y], color='red')

    axs.set_title('Convex Hull')

    # Show the plot
    plt.tight_layout()
    plt.show()
    
def convex_hull(points):
    lista = points.copy() 
    convex = [lista.pop(0)]
    for c in convex:
        pivot = c
        nex =  choice(lista)
        for index in lista:
            if index != nex:
                temp=[pivot, nex, index]
                if check_last3_points(temp) == False:
                    nex = index
        if nex not in convex:
            convex.append(nex)
            lista.remove(nex)
            visualization(points,convex)
        else:
            convex.append(points[0])
            return convex
                    
   
```
